```python
# Quantified outlier summary using IQR method
print('=== Outlier Count (IQR method: beyond Q3 + 1.5*IQR) ===')
for col in financial_cols:
    Q1 = application_train[col].quantile(0.25)
    Q3 = application_train[col].quantile(0.75)
    IQR = Q3 - Q1
    upper = Q3 + 1.5 * IQR
    lower = Q1 - 1.5 * IQR
    n_out = ((application_train[col] > upper) | (application_train[col] < lower)).sum()
    pct = 100 * n_out / len(application_train)
    print(f'  {col:<25s}: {n_out:>5,} outliers ({pct:.2f}%)')
```

> **Outlier Treatment Plan:**
> - `AMT_INCOME_TOTAL`: Log-transform before modelling. Winsorize at 99th percentile for distance-based models.
> - `AMT_CREDIT` / `AMT_GOODS_PRICE`: Tree-based models (LightGBM/XGBoost) are robust to these. No mandatory treatment.
> - `AMT_ANNUITY`: Flag high-annuity sub-group; engineer Annuity-to-Income ratio to normalise.
> - **Quantification matters** — knowing that X% of rows are outliers informs the imputation/capping decision.

### 5.3 Numerical Features vs TARGET (Boxplots)

```python
def bivariate_numerical(col, log_transform=False):
    """Boxplot of a numerical feature split by TARGET class."""
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    data = application_train.copy()
    xlabel = col
    if log_transform:
        data[col] = np.log1p(data[col])
        xlabel = f'log1p({col})'

    # Boxplot
    sns.boxplot(x='TARGET', y=col, data=data, palette=['#185FA5','#A32D2D'], ax=axes[0])
    axes[0].set_title(f'{xlabel} vs TARGET (Boxplot)', fontsize=11)
    axes[0].set_xticklabels(['No Default (0)', 'Default (1)'])

    # KDE overlay — BOTH classes on the SAME scale
    sns.kdeplot(data=data[data['TARGET']==0], x=col, label='No Default', fill=True,
                color='#185FA5', alpha=0.5, ax=axes[1])
    sns.kdeplot(data=data[data['TARGET']==1], x=col, label='Default', fill=True,
                color='#A32D2D', alpha=0.5, ax=axes[1])
    axes[1].set_title(f'{xlabel} by TARGET (KDE)', fontsize=11)
    axes[1].legend()

    # Statistical test
    g0 = data[data['TARGET']==0][col].dropna()
    g1 = data[data['TARGET']==1][col].dropna()
    stat, pval = stats.mannwhitneyu(g0, g1, alternative='two-sided')
    plt.suptitle(f'Mann–Whitney U p-value: {pval:.2e} — distributions are {"SIGNIFICANTLY" if pval<0.05 else "NOT significantly"} different',
                 fontsize=9, y=1.01)
    plt.tight_layout()
    plt.show()
```

## Bivariet analysis for categorical columns 

```python
def bivariate_categorical(col, figsize=(10, 5)):
    agg = application_train.groupby(col)['TARGET'].agg(['mean','count']).reset_index()
    agg.columns = [col, 'default_rate', 'n']
    agg = agg.sort_values('default_rate', ascending=False)

    fig, ax = plt.subplots(figsize=figsize)
    bars = ax.barh(agg[col].astype(str), agg['default_rate']*100,
                   color='#185FA5', edgecolor='white')
    ax.axvline(application_train['TARGET'].mean()*100, color='#A32D2D',
               linestyle='--', linewidth=1.2, label=f'Portfolio avg ({application_train["TARGET"].mean()*100:.1f}%)')

    for bar, (_, row) in zip(bars, agg.iterrows()):
        ax.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2,
                f'{row["default_rate"]*100:.1f}%  (n={row["n"]:,})',
                va='center', fontsize=8)

    ax.set_title(f'Default Rate by {col}', fontsize=12)
    ax.set_xlabel('Default Rate (%)')
    ax.legend(fontsize=9)
    plt.tight_layout()
    plt.show()
```

```python
#1. Credit to Income Ratio
application_train_fe['CREDIT_INCOME_RATIO'] = (
    application_train_fe['AMT_CREDIT'] /
    application_train_fe['AMT_INCOME_TOTAL']
)
# 2. Annuity to Income Ratio
application_train_fe['ANNUITY_INCOME_RATIO'] = (
    application_train_fe['AMT_ANNUITY'] /
    application_train_fe['AMT_INCOME_TOTAL']
)
# 3. Credit to Annuity Ratio
application_train_fe['CREDIT_ANNUITY_RATIO'] = (
    application_train_fe['AMT_CREDIT'] /
    application_train_fe['AMT_ANNUITY']
)
#4. Age in Years
application_train_fe['AGE_YEARS'] = (
    abs(application_train_fe['DAYS_BIRTH']) / 365
)
#5.Employment Years 
application_train_fe['EMPLOYMENT_YEARS'] = (
    abs(application_train_fe['DAYS_EMPLOYED']) / 365
)
#6. Income per Family Member
application_train_fe['INCOME_PER_FAMILY_MEMBER'] = (
    application_train_fe['AMT_INCOME_TOTAL'] /
    application_train_fe['CNT_FAM_MEMBERS']
)
#7.Children Ratio
application_train_fe['CHILDREN_RATIO'] = (
    application_train_fe['CNT_CHILDREN'] /
    application_train_fe['CNT_FAM_MEMBERS']
)
#8.Goods to Credit Ratio
application_train_fe['GOODS_CREDIT_RATIO'] = (
    application_train_fe['AMT_GOODS_PRICE'] /
    application_train_fe['AMT_CREDIT']
)
#9.External Sources Mean
application_train_fe['EXT_SOURCES_MEAN'] = (
    application_train_fe[
        [
            'EXT_SOURCE_1',
            'EXT_SOURCE_2',
            'EXT_SOURCE_3'
        ]
    ]
    .mean(axis=1)
)
#10.External Sources Standard Deviation
application_train_fe['EXT_SOURCES_STD'] = (
    application_train_fe[
        [
            'EXT_SOURCE_1',
            'EXT_SOURCE_2',
            'EXT_SOURCE_3'
        ]
    ]
    .std(axis=1)
)
#11.Employment to Age Ratio
application_train_fe['EMPLOYMENT_AGE_RATIO'] = (
    application_train_fe['EMPLOYMENT_YEARS'] /
    application_train_fe['AGE_YEARS']
)
```

## **Domain-Specific Feature Engineering**

To enhance the predictive power of our models beyond raw attributes, we engineered 11 domain-specific features. These features are grouped into critical banking and demographic risk indicators:

### **1. Financial Burden & Leverage Ratios**
These features evaluate the size of the loan relative to the client's income and their structural repayment capacity.
- **`CREDIT_INCOME_RATIO`** ($AMT\_CREDIT / AMT\_INCOME\_TOTAL$): Measures credit leverage. A high ratio indicates that the requested loan is significantly larger than the client's annual earnings, escalating risk.
- **`ANNUITY_INCOME_RATIO`** ($AMT\_ANNUITY / AMT\_INCOME\_TOTAL$): Represents the **Debt-to-Income (DTI)** ratio. It tracks the exact percentage of monthly/annual income consumed by the loan installment.
- **`CREDIT_ANNUITY_RATIO`** ($AMT\_CREDIT / AMT\_ANNUITY$): Indicates the theoretical duration or speed of loan repayment.

### **2. Socio-Demographic & Stability Indicators**
These features transform raw timeline counts into realistic, normalized scales for age, career stability, and household dependency.
- **`AGE_YEARS`**: Converts the raw `DAYS_BIRTH` timeline into actual age in years for better interpretability.
- **`EMPLOYMENT_YEARS`**: Converts raw `DAYS_EMPLOYED` into professional experience years.
- **`EMPLOYMENT_AGE_RATIO`**: Measures the percentage of a client's adult life spent in active employment—a highly powerful proxy for career consistency and reliable risk scoring.
- **`INCOME_PER_FAMILY_MEMBER`**: Captures real disposable income. A high gross income can be heavily diluted if split across a large household.
- **`CHILDREN_RATIO`**: Captures the household dependency burden by tracking the proportion of dependents (children) relative to the total family size.

### **3. Loan Intent & Collateral Assessment**
- **`GOODS_CREDIT_RATIO`** ($AMT\_GOODS\_PRICE / AMT\_CREDIT$): Compares the actual price of the goods being financed against the credit granted. This reflects whether the loan is fully collateralized or if the client is over-borrowing beyond the asset value.

### **4. Engineered Credit Scores (External Source Aggregations)**
Given that `EXT_SOURCE_1, 2, 3` are our strongest individual predictors, aggregating them extracts maximum cross-signal value.
- **`EXT_SOURCES_MEAN`**: Computes the baseline external credit health of the applicant. Averaging these scores smooths out individual missing components natively.
- **`EXT_SOURCES_STD`**: Measures the variance/discrepancy between different external rating agencies. High volatility across sources can point toward inconsistent credit histories.

### 7.1 Bureau Feature Aggregation
> **This was missing from the original notebook.** Raw bureau rows must be aggregated
> to one row per client before merging with the main dataset.

```python
bureau_agg = bureau.groupby('SK_ID_CURR').agg(
    BUREAU_LOAN_COUNT       = ('SK_ID_BUREAU',        'count'),
    BUREAU_ACTIVE_COUNT     = ('CREDIT_ACTIVE',        lambda x: (x == 'Active').sum()),
    BUREAU_CLOSED_COUNT     = ('CREDIT_ACTIVE',        lambda x: (x == 'Closed').sum()),
    BUREAU_MAX_OVERDUE      = ('CREDIT_DAY_OVERDUE',   'max'),
    BUREAU_MEAN_OVERDUE     = ('CREDIT_DAY_OVERDUE',   'mean'),
    BUREAU_TOTAL_DEBT       = ('AMT_CREDIT_SUM_DEBT',  'sum'),
    BUREAU_MAX_CREDIT       = ('AMT_CREDIT_SUM',       'max'),
    BUREAU_MEAN_PROLONGED   = ('CNT_CREDIT_PROLONG',   'mean'),
).reset_index()

bureau_agg['BUREAU_CLOSED_RATIO'] = (
    bureau_agg['BUREAU_CLOSED_COUNT'] / bureau_agg['BUREAU_LOAN_COUNT'].replace(0, np.nan)
)

print(f'Bureau aggregated shape: {bureau_agg.shape}')
bureau_agg.head()
```

```python
# Merge aggregated bureau features into main dataset
train_with_bureau = application_train_fe.merge(bureau_agg, on='SK_ID_CURR', how='left')
print(f'Shape after bureau merge: {train_with_bureau.shape}')

# Clients with no bureau history -> fill with 0 / meaningful defaults
bureau_feat_cols = bureau_agg.columns.drop('SK_ID_CURR').tolist()
no_bureau = train_with_bureau['BUREAU_LOAN_COUNT'].isna().sum()
print(f'Clients with no bureau history: {no_bureau:,} ({100*no_bureau/len(train_with_bureau):.1f}%)')
train_with_bureau[bureau_feat_cols] = train_with_bureau[bureau_feat_cols].fillna(0)
```

```python
# Validate: do bureau features correlate with TARGET?
bureau_corr = (
    train_with_bureau[bureau_feat_cols + ['TARGET']]
    .corrwith(train_with_bureau['TARGET'])
    .drop('TARGET')
    .abs()
    .sort_values(ascending=False)
)

fig, ax = plt.subplots(figsize=(9, 4))
bureau_corr.sort_values().plot(kind='barh', color='#534AB7', edgecolor='white', ax=ax)
ax.set_title('Bureau Aggregated Features — |Correlation| with TARGET', fontsize=11)
ax.set_xlabel('|Pearson Correlation|')
plt.tight_layout()
plt.show()
```
